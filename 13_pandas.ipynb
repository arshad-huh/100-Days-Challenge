```python
import numpy as np
import pandas as pd
```

```python
data =[
[1,'John',28,'New York',70000],
[2,'Anna',22,'Los Angeles',80000],
[3,'Peter',34,'Chicago',60000],
[4,'Linda',29,'Houston',72000],
[5,'James',45,'Phoenix',90000],
[6,'Laura',32,'San Antonio',85000],
[7,'Robert',38,'Dallas',67000],
[8,'Patricia',27,'San Diego',71000],
[9,'Michael',30,'San Jose',75000],
[10,'Mary',25,'Austin',78000]
        ]

col = ['id','name','age','city','salary']

df = pd.DataFrame(data, columns=col)

df
```

```python
df.set_index('id')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include=[object])
```

Find the mean age of employees.

```python
df['age'].mean()
```

Add a new column 'bonus' which is 10% of the 'salary'.

```python
df['bonus'] = df['salary'] * 0.10
df
```

Group the DataFrame by 'city' and find the average salary per city.

```python
df.groupby('city')['salary'].mean()
```

Sort the DataFrame by 'age' in descending order.

```python
df.sort_values('age', ascending=False)
```

Check for any missing values in the DataFrame.

```python
df.isnull().sum()
```

Calculate the rolling mean of the 'salary' column with a window size of 3.

```python
df['salary'].rolling(window=3).mean()
```

Find the employee with the second highest salary.

```python
df.nlargest(2, 'salary').iloc[1]
```

Replace missing values in the 'salary' column with the mean salary. Assume some missing values for this example.

```python
df.loc[3, 'salary'] = None
df['salary'].fillna(df['salary'].mean())
```

Calculate the correlation matrix of the numerical columns in the DataFrame.

```python
df.columns
```

```python
df[['id', 'age', 'salary']].corr()
```

Transform the 'name' column to uppercase.

```python
df['name'] = df['name'].str.upper()
df
```

Extract rows where the 'age' is an even number.

```python
df[df['age'] % 2 == 0]
```

Create a new DataFrame that contains only the employees whose salary is above the average salary.

```python
df[df['salary'] > df['salary'].mean()]
```

Create a DataFrame containing the count of employees in each city.

```python
city_employee_count = df['city'].value_counts().reset_index()
city_employee_count.columns = ['city', 'employee_count']
city_employee_count
```

Use the apply function to create a new column 'seniority' based on the 'age' column: "Senior" if age > 35, otherwise "Junior".

```python
df['seniority'] = df['age'].apply(lambda x: 'Senior' if x > 35 else 'Junior')
df
```
