```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Using the fmri dataset from seaborn, create a line plot to show the average signal for each timepoint, separated by event and region.

```python
fmri = sns.load_dataset('fmri')

sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region', markers=True)
plt.title('Average Signal by Timepoint, Event, and Region')
plt.show()
```

Using the planets dataset from seaborn, create a histogram to show the distribution of orbital periods, applying a logarithmic scale to the x-axis.

```python
planets = sns.load_dataset('planets')

plt.hist(planets['orbital_period'].dropna(), bins=50, log=True)

plt.title('Distribution of Orbital Periods (Log Scale)')
plt.show()
```

Using the exercise dataset from seaborn, create a facet grid of bar plots showing the average pulse by diet for each kind of activity.

```python
exercise = sns.load_dataset('exercise')

sns.catplot(kind='bar', x='diet', y='pulse', data=exercise, col='kind')
plt.show()
```
