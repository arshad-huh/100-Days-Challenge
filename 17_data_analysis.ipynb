```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the "tips" dataset from Seaborn and display the first 10 rows. What are the columns present in the dataset?

```python
tips = sns.load_dataset('tips')
print(tips.head(10))
print(tips.columns)
```

Calculate and display the summary statistics (mean, median, standard deviation, etc.) for the numerical columns in the "tips" dataset.

```python
tips.describe()
```

Create a histogram of the total bill amount. Use Seaborn to overlay a Kernel Density Estimate (KDE) plot on the histogram.



```python
sns.histplot(data=tips, x='total_bill', bins=30, kde=True)
plt.show()
```

Add a new column to the dataset that represents the tip as a percentage of the total bill. Display the first 5 rows of the modified dataset.

```python
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) *100
tips.head()
```

Calculate the average total bill, tip amount, and tip percentage for each day of the week. Display the results in a Pandas DataFrame.

```python
tips.groupby('day')[['total_bill', 'tip', 'tip_pct']].mean()
```

Create a scatter plot showing the relationship between total bill and tip amount. Use different colors to distinguish between different times of day (Lunch and Dinner).

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
```

 Create a box plot to compare the total bill amounts for different days of the week. What can you infer from the box plot?

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

Compute and display a heatmap of the correlation matrix for the numerical columns in the dataset. Which pairs of variables have the strongest correlations?

```python
corr = tips[['total_bill',	'tip'	,'size','tip_pct']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

Use a bar plot to show the average tip amount given by males and females. Which gender tends to tip more on average?

```python
sns.barplot(data=tips, x='sex', y='tip')
plt.show()
```

Create a pair plot for the dataset, color-coded by the 'smoker' column. What insights can you gain from the pair plot regarding the relationship between different variables?

```python
sns.pairplot(data=tips, hue='smoker')
plt.show()
```
