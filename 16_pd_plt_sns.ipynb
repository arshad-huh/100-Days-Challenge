```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Question 1: Create a violin plot showing the distribution of total_bill for each day in the tips dataset, and overlay a swarm plot to show individual data points.

```python
tips = sns.load_dataset('tips')

sns.violinplot(x='day', y='total_bill', data=tips, fill=False)
sns.swarmplot(x='day', y='total_bill', data=tips, alpha=0.5)
plt.show()
```

Question 2: Use the facetgrid method to create a grid of scatter plots showing sepal_length vs sepal_width for each species in the iris dataset.

```python
iris = sns.load_dataset('iris')

sns.relplot(kind='scatter', x='sepal_length', y='sepal_width', data=iris, col='species')
plt.show()
```

Question 4: Plot the distribution of total_bill in the tips dataset, separated by time (Lunch/Dinner), using a KDE plot.

```python
tips = sns.load_dataset('tips')

sns.kdeplot(x='total_bill', data=tips, hue="time", fill=True)
plt.show()
```

Question 5: Calculate and plot the average tip as a percentage of total_bill for each day in the tips dataset using a bar plot.

```python
aver = tips.groupby("day")['total_bill'].mean().reset_index()
aver['pct']= aver['total_bill'] / aver["total_bill"].sum() *100

sns.barplot(data=aver, y='day', x='pct')
plt.grid()
```

Question 6: Create a joint plot with regression line of body_mass_g vs flipper_length_mm in the penguins dataset, and show the marginal distributions.

```python
penguins = sns.load_dataset('penguins')
sns.jointplot(data=penguins, x='body_mass_g', y='flipper_length_mm', kind='reg', marginal_kws=dict(bins=15, fill=True))
plt.show()
```
